```python
import pandas as pd
```

```python
df = pd.read_csv('StudentsPerformance (1).csv')
print(df.info())
print(df[['gender', 'math score']])
print(df.describe())
```

```python
reading_scores = df['reading score'] > 70
print(df.value_counts())
print(len(reading_scores))
```

```python
# compute modes for each column (returns all modes; may have multiple rows if multimodal)
modes = df.mode()
print("Modes for each column (all):")
print(modes)

# single mode per column (take the first mode when multiple exist)
print("\nSingle mode per column:")
print(modes.iloc[0])

# mode for the boolean Series reading_scores
print("\nMode of reading_scores:")
print(reading_scores.mode().tolist())
```

```python
print(df.groupby(by = 'gender')['math score'].mean())
```

```python
print(df.groupby(by = 'gender')['math score'].agg(['min', 'max']))
```

```python
df.groupby(by = ['gender', 'parental level of education'])['writing score'].agg(['min', 'max', 'mean', 'median'])
```
